Syntax for how to use the API:
How to use: get_data(ticker, start_date = None, end_date = None, index_as_date = True, interval = “1d”)

```python
# Test run one ticker
amazon_weekly= get_data("amzn", start_date="12/04/2009", end_date="12/04/2021", index_as_date = True, interval="1wk")
```

Test run with multiple tickers in a list

```python
tickers = ["AMZN", "DIS", "BA"]
hist_data_dict = {}
for ticker in tickers:
    hist_data_dict[ticker] = get_data(ticker, start_date="12/04/2009", end_date="12/04/2021", index_as_date = True, interval="1wk")
```
